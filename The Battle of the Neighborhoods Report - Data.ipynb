```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_communities_in_Dubai"
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html5lib")  # create a soup object using the variable 'data'
```

```python
Dubai_Neighbors = pd.DataFrame(columns=["Community_Number", "Community_Name", "Community_Population"])
table=soup.find('table')
for row in table.findAll('tr'):
    col = row.find_all("td")
    if (col != []):
        CommunityNumber = col[0].text.strip() # .strip() used to remove any tag comes with the text
        CommunityName = col[1].text.strip() # .strip() used to remove any tag comes with the text
        CommunityPopulation = col[4].text.strip() # .strip() used to remove any tag comes with the text
        if(CommunityNumber !="" and CommunityName != "" and CommunityPopulation!=""):
            Dubai_Neighbors = Dubai_Neighbors.append({"Community_Number":CommunityNumber,
                                                      "Community_Name":CommunityName,
                                                      "Community_Population":CommunityPopulation}, ignore_index=True)
pd.set_option('display.max_rows', None)
Dubai_Neighbors
```
